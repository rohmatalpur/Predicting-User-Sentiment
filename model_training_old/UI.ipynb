```python
import tkinter as tk
from tkinter import ttk
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model
import pickle

# Load the pre-trained model, tokenizer, label encoder, and tone encoder
model = load_model('./sentiment_model.h5')  # Replace with the actual path to your model
with open('tokenizer.pkl', 'rb') as handle:  # Replace with the actual tokenizer file path
    tokenizer = pickle.load(handle)

with open('label_encoder.pkl', 'rb') as handle:  # Replace with the actual label encoder file path
    label_encoder = pickle.load(handle)

with open('tone_encoder.pkl', 'rb') as handle:  # Replace with the actual tone encoder file path
    tone_encoder = pickle.load(handle)

# Define a function to get prediction
def predict_sentiment(text, user_name):
    # Tokenize and pad the input text
    text_seq = tokenizer.texts_to_sequences([text])
    
    # Use the correct max_sequence_length (e.g., 63, if that's what was used in training)
    text_padded = pad_sequences(text_seq, maxlen=63)  # Adjust the maxlen to the correct value

    # Encode the user_name using the label encoder
    user_encoded = label_encoder.transform([user_name])

    # Predict the sentiment
    pred_prob = model.predict([text_padded, user_encoded])
    pred_label = np.argmax(pred_prob, axis=1)
    
    # Decode the prediction back to the sentiment class
    predicted_sentiment = tone_encoder.inverse_transform(pred_label)

    return predicted_sentiment[0]

# Create the main window
root = tk.Tk()
root.title("Twitter-like Sentiment Prediction")
root.geometry("400x550")
root.config(bg="#F5F8FA")  # Set background color similar to Twitter

# Define the available user names
user_names = [
    'lionel messi', 'HillaryClinton', 'realDonaldTrump', 'meghanmarkle',
    'Kenye West', 'SannaMarin', 'ImranKhanPTI', 'Joe Biden', 'Elon Musk',
    'Mr. Tweet', 'bolsonaro', 'benhamner', 'driltweets', 'cristiano ronaldo'
]

# Function to display the result
def on_predict_click():
    tweet = tweet_entry.get()  # Get tweet input
    user_name = user_name_combobox.get()  # Get user name from dropdown
    
    if tweet and user_name:
        predicted_sentiment = predict_sentiment(tweet, user_name)
        result_label.config(text=f"Predicted Sentiment: {predicted_sentiment}")
    else:
        result_label.config(text="Please enter both tweet and user name")

# Twitter-like top bar
top_bar = tk.Frame(root, bg="#1DA1F2", height=50)
top_bar.pack(fill="x", side="top")

# Title in the top bar
title_label = tk.Label(top_bar, text="Tweet Sentiment Prediction", fg="white", font=("Helvetica", 16), bg="#1DA1F2")
title_label.pack(pady=10)

# Frame for tweet input section
tweet_frame = tk.Frame(root, bg="#FFFFFF", pady=20)
tweet_frame.pack(pady=10, padx=10, fill="both", expand=True)

# Tweet input label
tweet_label = tk.Label(tweet_frame, text="What's happening?", font=("Helvetica", 14), fg="#1DA1F2", anchor="w", bg="#FFFFFF")
tweet_label.pack(fill="x", padx=10)

# Text box for tweet input
tweet_entry = tk.Entry(tweet_frame, width=50, font=("Helvetica", 12), bd=2, relief="solid", fg="#14171A")
tweet_entry.pack(pady=5, padx=10)

# User selection label
user_name_label = tk.Label(tweet_frame, text="Select User:", font=("Helvetica", 12), fg="#1DA1F2", anchor="w", bg="#FFFFFF")
user_name_label.pack(fill="x", padx=10, pady=5)

# Dropdown for selecting the user name
user_name_combobox = ttk.Combobox(tweet_frame, values=user_names, state="readonly", font=("Helvetica", 12))
user_name_combobox.pack(pady=5, padx=10)

# Button to predict the sentiment
predict_button = tk.Button(tweet_frame, text="Predict Sentiment", command=on_predict_click, font=("Helvetica", 14), bg="#1DA1F2", fg="white", bd=0, relief="solid")
predict_button.pack(pady=15, padx=10, fill="x")

# Frame for result display
result_frame = tk.Frame(root, bg="#FFFFFF", pady=20)
result_frame.pack(pady=10, padx=10, fill="both", expand=True)

# Label to display the prediction result
result_label = tk.Label(result_frame, text="Predicted Sentiment: ", font=("Helvetica", 14), fg="#14171A", anchor="w", bg="#FFFFFF")
result_label.pack(padx=10)

# Run the Tkinter event loop
root.mainloop()
```
