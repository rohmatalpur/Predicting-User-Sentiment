```python
# Import required libraries
import pandas as pd

# List all the CSV files to be merged
file_paths = [
    r'E:\AI project\updated tweets\messi_tweetsupdated.csv',
    r'E:\AI project\updated tweets\tweets.csv',
    r'E:\AI project\updated tweets\meghanmarkle tweets.csv',
    r'E:\AI project\updated tweets\kanye west tweets.csv',
    r'E:\AI project\updated tweets\SannaMarin.csv',
    r'E:\AI project\updated tweets\ImranKhanPTI.csv',
    r'E:\AI project\updated tweets\JoeBidenTweets.csv',
    r'E:\AI project\updated tweets\elon_musk_tweets.csv',
    r'E:\AI project\updated tweets\bolsonaro_tweetso.csv',
    r'E:\AI project\updated tweets\benhamner.csv',
    r'E:\AI project\updated tweets\driltweets.csv',
    r'E:\AI project\updated tweets\cr_tweetsupdated.csv'
]

# Create an empty list to store dataframes
dataframes = []

# Read and append each dataframe to the list
for file in file_paths:
    try:
        # Handling encoding errors and skipping bad lines
        df = pd.read_csv(file, encoding='utf-8', on_bad_lines='skip')
        if not df.empty:
            dataframes.append(df)  # Only append non-empty dataframes
        else:
            print(f"Skipped empty file: {file}")
    except Exception as e:
        print(f"Error reading {file}: {e}")

# Merge all dataframes into a single dataframe
if dataframes:
    try:
        merged_df = pd.concat(dataframes, ignore_index=True)
        # Save the merged dataframe into a single CSV file
        output_path = r'E:\AI project\merged_tweets.csv'
        merged_df.to_csv(output_path, index=False, encoding='utf-8')
        print(f"All files merged successfully into '{output_path}'")
    except Exception as e:
        print(f"Error during merging or saving: {e}")
else:
    print("No dataframes to merge.")
```

```python
#for identifying the tone of the tweets 

from textblob import TextBlob
import pandas as pd

# File paths
input_file_path = r"E:\AI project\filtered_tweets.csv"  # Path to the filtered CSV file
output_file_path = r"E:\AI project\tweets_with_tone.csv"  # Path to save the updated file

# Read the filtered CSV file
data = pd.read_csv(input_file_path)

# Add a new column 'tone' based on the sentiment analysis
def determine_tone(text):
    try:
        # Analyze the sentiment polarity
        polarity = TextBlob(text).sentiment.polarity
        return "positive" if polarity > 0 else "negative" if polarity < 0 else "neutral"
    except Exception as e:
        return "unknown"  # Handle any unexpected errors

data["tone"] = data["text"].apply(determine_tone)

# Save the updated DataFrame
data.to_csv(output_file_path, index=False)
print(f"Updated CSV with tone analysis saved at: {output_file_path}")
```

```python
#additional preprocessing to clean the tweet text by removing URLS, mentions , hashtags,. special characters, extra white spaces , convert to lower case

import pandas as pd
import re

# File paths
input_file_path = r"E:\AI project\tweets_with_tone.csv"  # Path to the input CSV file
output_file_path = r"E:\AI project\cleaned_tweets.csv"  # Path to save the cleaned file

# Read the CSV file
data = pd.read_csv(input_file_path)

# Make a copy of the DataFrame to preserve other columns' integrity
data_cleaned = data.copy()

# Define a function to preprocess the text
def preprocess_text(text):
    if isinstance(text, str):
        # Remove URLs
        text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)
        # Remove mentions (@username)
        text = re.sub(r"@\w+", "", text)
        # Remove hashtags (but keep the text, e.g., "#love" -> "love")
        text = re.sub(r"#", "", text)
        # Remove special characters and numbers
        text = re.sub(r"[^A-Za-z\s]", "", text)
        # Remove extra whitespaces
        text = re.sub(r"\s+", " ", text).strip()
        # Convert to lowercase
        text = text.lower()
        return text
    else:
        return text  # If it's not a string, return as is

# Apply the preprocessing function only to the 'text' column in the copy
data_cleaned["text"] = data_cleaned["text"].apply(preprocess_text)

# Save the cleaned DataFrame without affecting the original data
data_cleaned.to_csv(output_file_path, index=False)

print(f"Cleaned CSV saved at: {output_file_path}")
```
